```python
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.xlim(0, 0.1)

plt.xlabel('$\dot{\gamma}$', fontsize=25)
plt.ylabel('$(\eta_\mathrm{eff} - \eta_\mathrm{f}) / \phi_\mathrm{p}$', fontsize=25)

plt.plot(gammadot[1:], (viscosity_map['00'][1:] - MU) / PHI, label='$B_1=0  \ \ \ \ , \, B_2=\ \ \ \ 0$')
plt.plot(gammadot[1:], (viscosity_map['01'][1:] - MU) / PHI, label='$B_1=0.01, B_2=\ \ \,50.0$')
plt.plot(gammadot[1:], (viscosity_map['02'][1:] - MU) / PHI, label='$B_1=0.01, B_2=     -50.0$')
plt.legend(fontsize=18)
```
