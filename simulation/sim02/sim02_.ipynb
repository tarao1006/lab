```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('presentation_')
with open('data.pickle', 'rb') as f:
    data = pickle.load(f)
degrees = [(2 * np.pi) * d / 32 for d in range(0, 33)]
```

```python
xticks = np.linspace(0, 2 * np.pi, 5, dtype=np.float64)
xticks_str = ['$0$', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']
yticks = np.linspace(-20, 20, 5, dtype=np.int8)

plt.xticks(xticks, xticks_str, fontsize=50)
plt.yticks(yticks, fontsize=50)
plt.ylim(-20.0, 20.0)
plt.xlabel(r'$\it{\theta}$', fontsize=50)
plt.ylabel(r'$\it{\Sigma^\mathrm{(p)}_{xy}}_{\times\rm{10^5}}$', fontsize=50)

plt.plot(degrees, data['calculated_shear_stress'] * 10 ** 5, linewidth=3)

plt.tight_layout()
plt.savefig('../figs/sigma_xy.png')
```

```python
fig, ax = plt.subplots()

xticks = np.linspace(0, 2 * np.pi, 5, dtype=np.float64)
xticks_str = ['$0$', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']
yticks = np.linspace(-20, 20, 5, dtype=np.int8)

plt.xticks(xticks, xticks_str, fontsize=50)
plt.yticks(yticks, fontsize=50)
plt.ylim(-20.0, 20.0)
plt.xlabel(r'$\it{\theta}$', fontsize=50)
plt.ylabel(r'$\it{\Sigma^\mathrm{(p)}_{xy}}_{\times\rm{10^5}}$', fontsize=50)

ax.plot(degrees, data['calculated_shear_stress'] * 10 ** 5, linewidth=3, color="k")
ax.axvspan(np.pi * 0/2, np.pi * 1/2, facecolor='lightskyblue', lw=0)
ax.axvspan(np.pi * 1/2, np.pi * 2/2, facecolor='wheat', lw=0)
ax.axvspan(np.pi * 2/2, np.pi * 3/2, facecolor='lightskyblue', lw=0)
ax.axvspan(np.pi * 3/2, np.pi * 4/2, facecolor='wheat', lw=0)


plt.tight_layout()
plt.savefig('../figs/sigma_xy.png')
```
